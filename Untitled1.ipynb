```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("train.csv")
data.head()
```

```python
data.columns
```

```python
plt.hist(data['Age'])
plt.xlabel('Age')
plt.ylabel('Number of passengers')
plt.title('Distribution of age in dataset')
plt.show()
```

```python

plt.figure(figsize=(16,8))
axis = sns.kdeplot(data.Age[data.Survived == 0], label="died")
axis = sns.kdeplot(data.Age[data.Survived == 1], label="survived")
plt.title("Survival with Age")
plt.xlabel("Age")
plt.ylabel("count")
plt.show()
```

```python
plt.hist(data['Fare'])
plt.xlabel('Fare')
plt.ylabel('Number of passengers')
plt.title('Distribution of fare paid in dataset')
plt.show()
```

```python
sns.countplot(data['Survived'])
plt.title("NOT SURVIVED AND SURVIVED")
plt.xlabel("Survived/didnot")
plt.ylabel('Count')
plt.show()
```

```python
sex_count_values =data['Sex'].value_counts()
sex_count_values
```

```python
plt.bar(data['Sex'].value_counts().index, data['Sex'].value_counts().values)
plt.xlabel('Sex')
plt.ylabel('Number of passengers')
plt.title('Sex distribution')
plt.show()
```

```python
male_dead =((data['Sex'] =='male') & (data['Survived'] ==0)).sum()
print("dead male", male_dead)
male_survived =((data['Sex'] =='male') & (data['Survived'] ==1)).sum()
print("alive male",male_survived)
```

```python
plt.bar(['DEAD','SURVIVED'],[male_dead,male_survived])
plt.xlabel("survival status")
plt.ylabel("NUmber of male passengeers")
plt.title("number of dead and survived malepassengers")
plt.show()
```

```python
# assuming its  A client and the client has asked you to check which  factor affected  the survival of a passenger most so that means that the target row is  survived  so we can compare a bunch of other factors with survival to see which one suits best to like train the model to get the best accuracy
first_class_count = (data['Pclass'] == 1).sum()
print("First class :", first_class_count)
second_class_count = (data['Pclass'] == 2).sum()
print("second class :", second_class_count)
third_class_count = (data['Pclass'] == 3).sum()
print("third class :", third_class_count)

```

```python
labels = ['First class', 'second class', 'Third class']
sizes = [first_class_count, second_class_count, third_class_count]
plt.pie(sizes,labels =labels ,autopct='%1.2f%%')
plt.title("class distribution")
plt.show()
```

```python
# assuming its  A client and the client has asked you to check which  factor affected  the survival of a passenger most so that means that the target row is  survived  so we can compare a bunch of other factors with survival to see which one suits best to like train the model to get the best accuracy
sns.countplot(x=data['Pclass'], hue=data['Survived'])
plt.title("survival as per class")
plt.xlabel("classes")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(16,8))
axis = sns.kdeplot(data.Pclass[data.Survived == 0], label="died")
axis = sns.kdeplot(data.Pclass[data.Survived == 1], label="survived")
plt.title("Survival with Pclass")
plt.xlabel("Pclass")
plt.ylabel("count")
plt.show()
```
